## 불러오기(import)

기본 import:


````python
# math.sqrt(16) # import 되어 있지 않음
````

````python
import math
````

````python
print(math.sqrt(16))  # math 모듈의 sqrt 함수 사용
````

별칭으로 import:



````python
import math as m
print(m.sqrt(16))  # 별칭 m을 사용하여 math 모듈의 sqrt 함수 호출
````

특정 함수 또는 클래스만 import:



````python
from math import sqrt
print(sqrt(16))  # 직접 sqrt 함수 호출
````
